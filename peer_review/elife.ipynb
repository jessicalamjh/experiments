```python
from glob import glob
from lxml import etree
import os.path as path

data_dir = "data/elife"
```

```python
filepaths = sorted(list(glob(f"{data_dir}/articles/*.xml")))
```

```python
for filepath in filepaths[-100:]:
    tree = etree.parse(filepath)
    root = tree.getroot()

    for _ in root.findall(".//article-title"):
        print(path.basename(filepath))
        print(_.text)
        print()
```
